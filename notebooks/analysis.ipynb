```python
import sys; sys.path.insert(0, '..')
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from src.black_scholes import BlackScholes
```

```python
bs = BlackScholes(S=150, K=155, T=0.5, r=0.05, sigma=0.25)
print(f"Call Price: ${bs.call_price():.2f}")
print(f"Put Price:  ${bs.put_price():.2f}")
```

```python
greeks = bs.all_greeks('call')
df = pd.DataFrame(greeks.items(), columns=['Greek', 'Value'])
print(df.to_string(index=False))
```

```python
stock_prices = np.linspace(100, 200, 300)
calls = [BlackScholes(s, 155, 0.5, 0.05, 0.25).call_price() for s in stock_prices]
puts  = [BlackScholes(s, 155, 0.5, 0.05, 0.25).put_price()  for s in stock_prices]
plt.figure(figsize=(10,5))
plt.plot(stock_prices, calls, label='Call', color='green')
plt.plot(stock_prices, puts,  label='Put',  color='red')
plt.axvline(155, linestyle='--', color='gray', label='Strike K=155')
plt.legend(); plt.grid(True, alpha=0.3)
plt.title('BS Option Prices vs Stock Price')
plt.xlabel('Stock Price ($)'); plt.ylabel('Option Price ($)')
plt.show()
```
